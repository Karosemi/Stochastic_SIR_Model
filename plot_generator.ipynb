```python
import os
import numpy as np
from numpy import mean
import matplotlib.pyplot as plt
from multiprocessing import Pool
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import pickle
```

```python
def open_pickle_file(filename):
    with open(filename, 'rb') as file:
        try:
            return pickle.load(file)
        except Exception:
            print(f"pickle.load of {filename} ended with an exception")
            raise
```

```python

pkl_ratio_files = ["L_16.pkl", "L_32.pkl","L_64.pkl","L_128.pkl","L_256.pkl","L_512.pkl"]
```

```python
plt.style.use('ggplot')
```

```python
step = 20
cs = np.linspace(0.15, 0.2, step)
plt.figure(figsize=(18,12))
for pkl_ratio_file in pkl_ratio_files:
    L_array = open_pickle_file(pkl_ratio_file)
    U = L_array[2]
    Ro = L_array[3]
    up = U*Ro
    plt.plot(cs, U, 'o-', label=pkl_ratio_file.replace("_","=" ).replace(".pkl", ""))
plt.legend()
plt.xlabel('c')
plt.ylabel('U')
plt.savefig("U.png")
plt.show()
```
